## Bot Contra Fatos

```python
cfatos_urls = []
for pagina in range(0, 6):
  urls = (f"https://www.contrafatos.com.br/politica/page/{pagina}")
  cfatos_urls.append(urls)
```

```python
artigos = []
for paginas in cfatos_urls:
  time.sleep(5)
  requisicao = requests.get(paginas)
  soup = BeautifulSoup(requisicao.text)
  for x in soup:
    blocos_noticias = soup.find_all('div', {'class':'zox-main-blog zoxrel left zox100'})
    for bloco in blocos_noticias:
      url = bloco.find('a')['href']
      titulo = bloco.find('h2', {'class':'zox-s-title2'}).text
      data = bloco.find('span', {'class':'zox-byline-date'}).text
      artigos.append([titulo, data, url])
```

```python
for paginas in cfatos_urls:
  time.sleep(5)
  requisicao = requests.get(paginas)
  soup = BeautifulSoup(requisicao.text)
  for x in soup:
    blocos_noticias = soup.find_all('div', {'class':'zox-widget-flex1-wrap zox-widget-flex1-col zox-div3'})
    for bloco in blocos_noticias:
      url = bloco.find('a')['href']
      titulo = bloco.find('h2', {'class':'zox-s-title2'}).text
      data = bloco.find('span', {'class':'zox-byline-date'}).text
      artigos.append([titulo, data, url])
```

```python
dados_contra_fatos = pd.DataFrame(artigos, columns=['Titulo', 'Data', 'Link'])
dados_contra_fatos
```

```python
dados_contra_fatos['Entidades_Nomeadas'] = dados_contra_fatos['Titulo'].apply(lambda text: [ent.text for ent in nlp(text).ents])
print(dados_contra_fatos[['Entidades_Nomeadas', 'Data']])
```

## Bot [Aliados Brasil](https://www.aliadosbrasiloficial.com.br/)

```python
paginas_politica = []
for pagina in range(1, 6):
  urls = (f"https://www.aliadosbrasiloficial.com.br/noticias/politica&page={pagina}")
  paginas_politica.append(urls)
```

```python
artigos = []
for paginas in paginas_politica:
  time.sleep(5)
  requisicao = requests.get(paginas)
  html = BeautifulSoup(requisicao.text)
  for noticia in html:
    politica_aliados = html.find('div', {'list_posts'})
    aliados_politica = politica_aliados.findAll('a')
    for noticia in aliados_politica:
      link = noticia.get('href')
      print(link)
      artigos.append(link)
```

```python
todos_artigos = [item for item in artigos if item not in paginas_politica]
```

```python
artigos_sem_duplicatas = set(todos_artigos)
todos_artigos = list(artigos_sem_duplicatas)
```

```python
todos_artigos.remove('https://www.aliadosbrasiloficial.com.br/noticias/politica&page=30')
```

```python
todos_artigos.remove('https://www.aliadosbrasiloficial.com.br/noticias/politica&page=6')
```

```python
todos_artigos
```

```python
lista_textos = []
for texto in todos_artigos:
  time.sleep(2)
  requisicao = requests.get(texto)
  html = BeautifulSoup(requisicao.text)
  for conteudo in html:
    titulo = html.find('h2').text
    data = html.find('div', {'post_info_date_post'}).text
    lista_textos.append([titulo, data, texto])
```

```python
dados_aliados = pd.DataFrame(lista_textos, columns=['Titulo', 'Data', 'Link'])
```

```python
dados_aliados.drop_duplicates(inplace=True)
```

```python
dados_aliados['Data'] = dados_aliados['Data'].str.strip()
dados_aliados['Data'] = pd.to_datetime(dados_aliados['Data'], format='%d/%m/%Y %H:%M')
dados_aliados['Data'] = dados_aliados['Data'].dt.strftime('%d/%m/%Y')
```

```python
dados_aliados
```

```python
dados_aliados['Entidades_Nomeadas'] = dados_aliados['Titulo'].apply(lambda text: [ent.text for ent in nlp(text).ents])
print(dados_aliados[['Entidades_Nomeadas', 'Data']])
```

```python
enti_aliados = pd.DataFrame(dados_aliados[['Entidades_Nomeadas', 'Data']])
enti_aliados
```

```python
enti_aliados.to_csv('Aliados.csv')
```
